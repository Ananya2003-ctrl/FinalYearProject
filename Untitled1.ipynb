```python
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/jigsaw-toxic-comment-train.csv")
df.shape
# X=df['comment_text']
# Y=df[df.columns[2:]]
```

```python
sample_size=50000
sampled_df=df.sample(n=sample_size, random_state=42)
X=sampled_df['comment_text']
Y=sampled_df[sampled_df.columns[2:]].values
```

```python
def tokenize_data(text, max_length=128):
    return tokenizer(text, padding='max_length', truncation=True, max_length=max_length, return_tensors='tf')
X=X.tolist()
tokenized_inputs=[tokenize_data(comment) for comment in X]
```

```python
from transformers import TFBertForSequenceClassification
import tensorflow as tf
model=TFBertForSequenceClassification.from_pretrained('bert-base-uncased',num_labels=6)
# numm_label referes to output shape 6 as we hae 6 lables
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
#tensorflow dataset
# 2 variables> input_ids--> varibles that conatins tokenized representation of input comments generated by bert tokenizer
# each token has a unique id used ny the vocabulary built
# attention_masks--> indicates which tokns are actual words and which are padding,ie tokens that fill space with 0 when comments is shorter than max_length
input_ids=tf.convert_to_tensor([input_data['input_ids'] for input_data in tokenized_inputs])
attention_masks=tf.convert_to_tensor([input_data['attention_mask']for input_data in tokenized_inputs])
#model expects 2d input with number of samples and sequence_length..but daat has 3 dimension..so squeeze to reduce dim
input_ids=tf.squeeze(input_ids)
attention_masks=tf.squeeze(attention_masks)
print(input_ids.shape)
print(attention_masks.shape)
Y=tf.convert_to_tensor(Y)
dataset_tensor=tf.data.Dataset.from_tensor_slices(({'input_ids':input_ids,'attention_mask': attention_masks},Y))
batch_size=16
dataset_tensor=dataset_tensor.shuffle(1000).batch(batch_size).prefetch(tf.data.AUTOTUNE)
```

```python
for batch in dataset_tensor.take(1):
    print("batch input shape:", batch[0]['input_ids'].shape, batch[0]['attention_mask'].shape )
```

```python
train=dataset_tensor.take(int(len(dataset_tensor)*0.7))
valid=dataset_tensor.skip(int(len(dataset_tensor)*0.7)).take(int(len(dataset_tensor)*0.2))
test=dataset_tensor.skip(int(len(dataset_tensor)*0.9)).take(int(len(dataset_tensor)*0.1))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
predictions=model.predict(test)
```

```python
test_pred_labels=tf.round(tf.nn.sigmoid(predictions['logits'])).numpy()
```

```python
Y_test=[]
for batch in test:
  labels=batch[1]
  Y_test.append(labels)
Y_test=tf.concat(Y_test,axis=0)
print(Y_test.shape)
```

```python
y_true=tf.reshape(Y_test,[-1])
y_pred=tf.reshape(test_pred_labels,[-1])
precision=precision_score(y_true,y_pred, average='weighted')
recall=recall_score(y_true,y_pred, average='weighted')
f1=f1_score(y_true,y_pred, average='weighted')
print(f"Precision:{precision}")
print(f"Recall:{recall}")
print(f"F1:{f1}")
print(classification_report(y_true,y_pred))
```
